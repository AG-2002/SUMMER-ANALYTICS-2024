### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Calculate median MPG
median_mpg = df['mpg'].median()
print(f"Median MPG: {median_mpg}")

# Calculate above-median MPG
above_median = df[df['mpg'] > median_mpg].copy()
print(f"Above-median MPG: {above_median}")

# Calculate horsepower-to-weight ratio

above_median['hp_to_weight'] = above_median['horsepower'] / above_median['weight']

#Find the Car(s) with the Highest Ratio
max_hp_to_weight = above_median['hp_to_weight'].max()
best_ratio_cars = above_median[above_median['hp_to_weight'] == max_hp_to_weight]

print(best_ratio_cars[['name']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# Calculate average mpg by year and origin
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Set the style
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Create the line plot
sns.lineplot(data=avg_mpg,
             x='model_year',
             y='mpg',
             hue='origin',
             palette={'usa': 'blue', 'japan': 'red', 'europe': 'green'},
             marker='o',
             linewidth=2.5)

# Customize the plot
plt.title('Evolution of Average MPG by Origin (1970-1982)', fontsize=14, pad=20)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.xticks(range(70, 83))  # Show all years from 70 to 82
plt.grid(True, alpha=0.3)

# Improve legend
plt.legend(title='Origin',
           loc='upper left',
           labels=['Japan', 'Europe','USA'])

# Show the plot
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
repeated_name = df.groupby(['name'])['model_year'].nunique()
repeated_names = np.array(repeated_name[repeated_name > 1].index)

df1 = df[df['name'].isin(repeated_names)]
lowest_var_name = df1.groupby('name')['mpg'].std().sort_values().index[0]
print("Lowest Variation in mpg with at least 2 appearances and std(mpg) < 1.0 is\t", lowest_var_name)
print("\n\n\n", df1.groupby('name').agg(
    appearances = ('name', 'count'),
     average_mpg = ('mpg', 'mean')).sort_values(by = 'appearances', ascending=False))

print("\n\n\n", df1.groupby('name').agg(
    appearances = ('name', 'count'),
     average_mpg = ('mpg', 'mean')).sort_values(by = 'average_mpg', ascending=False))
```
